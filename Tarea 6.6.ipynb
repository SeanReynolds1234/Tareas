### Tarea 6.6

Use el método de bisección para encontrar un cero de la función 

$$
f(x) = 1 - \cos(x) -\frac{\sin(x)}{x}
$$

entre $-6.5$ y $-6.0$. Grafique el error relativo requerido al calcular el cero en función del número de iteraciones que necesita el método para converger. Compare con la cota teórica.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Se define la función entregada
def f(x):
    return 1 - np.cos(x) - np.sin(x)/x
```

```python
x = np.linspace(-6.5, -6, 100) #Se crea un espacio para poder graficar esta función
plt.plot(x, f(x))
plt.show()
```

```python
#Se define el método de bisección
def biseccion(func, a, b, epsilon, N=1000000):
    """
    Método de bisección, que entrega un valor que toma una función 
    seleccionada, que al ser evaluado en esta misma entrega un número 
    muy cercano al cero de la función. Para lograr esto se deben de 
    especificar la función, dos valores iniciales donde el usuario cree 
    que puede encontrarse el cero y un error asociado al valor que te 
    entregue el método.
    
    En esta definición no solo entrega un valor aproximado para el cero
    de la función, sino que también entrega el número de pasos que se
    hicieron y el error permitido por el usuario.
    
    """
    #Utilizando un ciclo for, se pueden definir en qué momento el método se detiene
    for i in range(N):
        p = a + (b - a)/2 
        if abs(func(p)) < epsilon:
            return i, epsilon, (b-a)/2**i, p
        elif func(p)*func(a) > 0:
            a = p
        else:
            b = p
    #Error en caso de que el método no encuentre un valor en N=1000000 de pasos
    print("No se encontro el cero despues de", N, "iteraciones")
```

Una vez definido el método de bisección, se continúa creando los datos para poder comparar los errores en función de los pasos y la cota teórica, la cual es la siguiente:

$$
|p_n - p| \leq \frac{b - a}{2^n}
$$

donde $|p_n - p|$ se asocia al error y $n$ los pasos requeridos para llegar a un valor.

```python
#Pasos requeridos. Se toma la posición [0], debido a que esos son los pasos
pasos1 = biseccion(f, -6.3, -6.2, 1e-3)[0]
pasos2 = biseccion(f, -6.3, -6.2, 1e-4)[0]
pasos3 = biseccion(f, -6.3, -6.2, 1e-5)[0]
pasos4 = biseccion(f, -6.3, -6.2, 1e-6)[0]
pasos5 = biseccion(f, -6.3, -6.2, 1e-7)[0]
pasos6 = biseccion(f, -6.3, -6.2, 1e-8)[0]
pasos7 = biseccion(f, -6.3, -6.2, 1e-9)[0]
pasos8 = biseccion(f, -6.3, -6.2, 1e-10)[0]
pasos9 = biseccion(f, -6.3, -6.2, 1e-11)[0]
```

```python
#Errores relativos. Se toma la posición [1], debido a que esos son los errores
error1 = biseccion(f, -6.3, -6.2, 1e-3)[1]
error2 = biseccion(f, -6.3, -6.2, 1e-4)[1]
error3 = biseccion(f, -6.3, -6.2, 1e-5)[1]
error4 = biseccion(f, -6.3, -6.2, 1e-6)[1]
error5 = biseccion(f, -6.3, -6.2, 1e-7)[1]
error6 = biseccion(f, -6.3, -6.2, 1e-8)[1]
error7 = biseccion(f, -6.3, -6.2, 1e-9)[1]
error8 = biseccion(f, -6.3, -6.2, 1e-10)[1]
error9 = biseccion(f, -6.3, -6.2, 1e-11)[1]
```

```python
#Cota Teórica. Se toma la posición [2], debido a que esos son las cotas teóricas
cota_t1 = biseccion(f, -6.3, -6.2, 1e-3)[2]
cota_t2 = biseccion(f, -6.3, -6.2, 1e-4)[2]
cota_t3 = biseccion(f, -6.3, -6.2, 1e-5)[2]
cota_t4 = biseccion(f, -6.3, -6.2, 1e-6)[2]
cota_t5 = biseccion(f, -6.3, -6.2, 1e-7)[2]
cota_t6 = biseccion(f, -6.3, -6.2, 1e-8)[2]
cota_t7 = biseccion(f, -6.3, -6.2, 1e-9)[2]
cota_t8 = biseccion(f, -6.3, -6.2, 1e-10)[2]
cota_t9 = biseccion(f, -6.3, -6.2, 1e-11)[2]
```

Finalmente, se procede a realizar la graficas para así poder comparar los diferentes datos obtenidos. Para este caso, se irán evaluando diferentes grupos de errores, para ir viendo de a poco cómo se diferencian las gráficas entre el error y la cota teórica.

Gráficas para errores (epsilon) de: $10^{-9}$, $10^{-10}$ y $10^{-11}$

```python
#Gráfica de errores en función de los pasos
plt.plot(pasos7, error7, 'o', color='r')
plt.plot(pasos8, error8, 'o', color='r')
plt.plot(pasos9, error9, 'o', color='r')
plt.title('Errores vs Número de pasos')
plt.xlabel('Número de pasos')
plt.ylabel('Error relativo')
plt.show()
```

```python
#Gráfica de la cota teórica en función de los pasos
plt.plot(pasos7, cota_t7, 'o', color='r')
plt.plot(pasos8, cota_t8, 'o', color='r')
plt.plot(pasos9, cota_t9, 'o', color='r')
plt.title('Cota teórica vs Número de pasos')
plt.xlabel('Número de pasos')
plt.ylabel('Cota teórica')
plt.show()
```

Se observa una gran discrepancia entre los valores de laas gráficas, siendo los valores de la cota teórica bastante menores los del error relativo, por una diferencia del orden de $10^{-6}$.

Gráficas para errores (epsilon) de: $10^{-6}$, $10^{-7}$ y $10^{-8}$
